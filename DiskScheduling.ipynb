```python
def fcfs(nums, start):
    total = 0
    for x,y in enumerate(nums):
        if x == 0:
            total += abs(start-y)
        else:
            total += abs(nums[x-1] - y)
    return total
```

```python
def sstf(nums, start):
    arr = nums.copy()
    total = 0
    current = start
    while len(arr) > 0:
        minimum = float('inf')
        minNum = 0
        for x in arr:
            if abs(current-x) < minimum:
                minimum = abs(current-x)
                minNum = x
        current = minNum
        arr.remove(minNum)
        print(minNum)
        total += minimum
        
    return total
```

```python
def scan(nums, start, prev, numTracks):
    
    seek_count = 0
    distance, cur_track = 0, 0
    left = []
    right = []
    seek_sequence = []
    
    if start < prev:
        direction = "left"
    else:
        direction = "right"

    if (direction == "left"):
        left.append(0)
    elif (direction == "right"):
        right.append(numTracks - 1)

    for i in range(len(nums)):
        if (nums[i] < start):
            left.append(nums[i])
        if (nums[i] > start):
            right.append(nums[i])

    left.sort()
    right.sort()

    run = 2
    while (run != 0):
        if (direction == "left"):
            for i in range(len(left) - 1, -1, -1):
                cur_track = left[i]
                seek_sequence.append(cur_track)
                distance = abs(cur_track - start)
                seek_count += distance
                start = cur_track
            
            direction = "right"
    
        elif (direction == "right"):
            for i in range(len(right)):
                cur_track = right[i]
                seek_sequence.append(cur_track)
                distance = abs(cur_track - start)
                seek_count += distance
                start = cur_track
            
            direction = "left"
        
        run -= 1
    print(seek_sequence)
    return seek_count
```

```python
def look(nums, start, prev):
    
    seek_count = 0
    distance = 0
    cur_track = 0
    
    if start < prev:
        direction = "left"
    else:
        direction = "right"

    left = []
    right = []

    seek_sequence = []

    for i in range(len(nums)):
        if (nums[i] < start):
            left.append(nums[i])
        if (nums[i] > start):
            right.append(nums[i])

    left.sort()
    right.sort()

    run = 2
    while (run):
        if (direction == "left"):
            for i in range(len(left) - 1, -1, -1):
                cur_track = left[i]
                seek_sequence.append(cur_track)
                distance = abs(cur_track - start)
                seek_count += distance
                start = cur_track
            direction = "right"
            
        elif (direction == "right"):
            for i in range(len(right)):
                cur_track = right[i]
                seek_sequence.append(cur_track)
                distance = abs(cur_track - start)
                seek_count += distance
                start = cur_track
            direction = "left"
        run -= 1
    print(seek_sequence)
    return seek_count
```

```python
def cscan(nums, start, numTracks):

    seek_count = 0
    distance = 0
    cur_track = 0
    left = []
    right = []
    seek_sequence = []

    left.append(0)
    right.append(numTracks - 1)

    for i in range(len(nums)):
        if (nums[i] < start):
            left.append(nums[i])
        if (nums[i] > start):
            right.append(nums[i])
            
    left.sort()
    right.sort()

    for i in range(len(right)):
        cur_track = right[i]
        seek_sequence.append(cur_track)
        distance = abs(cur_track - start)
        seek_count += distance
        start = cur_track
    
    start = 0

    seek_count += (numTracks - 1)

    for i in range(len(left)):
        cur_track = left[i]
        seek_sequence.append(cur_track)
        distance = abs(cur_track - start)
        seek_count += distance
        start = cur_track
    return seek_count
```

```python
def clook(nums, start, numTracks):
    
    seek_count = 0
    distance = 0
    cur_track = 0
    left = []
    right = []
    seek_sequence = []


    for i in range(len(nums)):
        if (nums[i] < start):
            left.append(nums[i])
        if (nums[i] > start):
            right.append(nums[i])
            
    left.sort()
    right.sort()

    for i in range(len(right)):
        cur_track = right[i]
        seek_sequence.append(cur_track)
        distance = abs(cur_track - start)
        seek_count += distance
        start = cur_track
    
    seek_count += abs(start - left[0])
    start = left[0]

    for i in range(len(left)):
        cur_track = left[i]
        seek_sequence.append(cur_track)
        distance = abs(cur_track - start)
        seek_count += distance
        start = cur_track
    return seek_count
```

```python
test = [40, 48, 14, 44, 90,  7, 55]
start = 15
previous = 10
numTracks = 100

print("FCFS:",fcfs(test, start))
print("SSTF:",sstf(test, start))
print("SCAN:",scan(test, start, previous, numTracks))
print("LOOK:",look(test, start, previous))
print("C-SCAN:",cscan(test, start, numTracks))
print("C-LOOK:",clook(test, start, numTracks))
```
